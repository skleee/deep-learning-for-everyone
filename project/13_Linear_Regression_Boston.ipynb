```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.models import Sequential # 'tensorflow' 붙여야 함. version문제
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd
```

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/deep-learning-for-everyone/dataset/housing.csv", delim_whitespace=True, header=None)
df.head()
```

```python
df.info()
```

```python
dataset = df.values
X = dataset[:,0:13]
Y = dataset[:,13]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)

model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1)) # 출력층에 활성화 함수 X

model.compile(loss='mean_squared_error',
              optimizer='adam')

model.fit(X_train, Y_train, epochs=200, batch_size=10)
```

```python
Y_prediction = model.predict(X_test).flatten() # flatten(): 1차원으로

for i in range(10):
    label = Y_test[i]
    prediction = Y_prediction[i]
    print("실제가격: {:.3f}, 예상가격: {:.3f}".format(label, prediction))
```
