# Linear Regression

### Least Square Method
: x의 편차와 y의 편차를 곱해서 합한 값 / x의 편차의 제곱의 합

```python
import numpy as np
```

```python
# x, y
x = [2, 4, 6, 8]
y = [81, 93, 91, 97]

# x, y의 평균값
mx = np.mean(x)
my = np.mean(y)
print("Mean of x: ", mx)
print("Mean of y: ", my)
```

```python
# 기울기 공식의 분모
denominator = sum([(mx-i)**2 for i in x])

# 기울기 공식의 분자
def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx)*(y[i] - my)
    return d
numerator = top(x, mx, y, my)

print("분모: ", denominator)
print("분자: ", numerator)
```

```python
# 기울기 a 와 y절편 b 구하기
a = numerator / denominator
b = my - (mx*a)

print("a: ", a)
print("b: ", b)
```
