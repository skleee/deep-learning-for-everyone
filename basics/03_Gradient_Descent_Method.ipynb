```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.show()
```

```python
# 리스트로 되어 있는 x, y를 넘파이 배열로 변경 (인덱스로 하나씩 불러와 계산하기 위해)
x_data = np.array(x)
y_data = np.array(y)

# 기울기 a와 y 절편 b의 값 초기화
a = 0
b = 0

# learning rate
lr = 0.05 

# epoch
epochs = 2001 

# Gradient Descent Method
for i in range(epochs): # epoch 수 만큼 반복
    y_hat = a * x_data + b  # y
    error = y_data - y_hat  # 오차
    a_diff = -(1/len(x_data)) * sum(x_data * (error)) # 오차함수를 a에 대해 미분 
    b_diff = -(1/len(x_data)) * sum(error)  # 오차함수를 b에 대해 미분 
    a = a - lr * a_diff  # 학습률을 곱해 a 업데이트
    b = b - lr * b_diff  # 학습률을 곱해 b 업데이트
    if i % 100 == 0:
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))
```

```python
# 앞서 구한 a, b로 그래프 그림
y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```
