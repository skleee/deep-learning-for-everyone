```
import numpy as np

# 가상의 기울기 a와 y절편 b
fake_a_b = [3, 76]

# x, y
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

# y=ax + b에 a,b 대입하여 예측 값을 출력하는 함수
def predict(x):
    return fake_a_b[0]*x + fake_a_b[1]

# MSE를 구하는 함수 (각 (예측 값 - 실제 값)의 제곱의 평균)
def mse_val(predict_result,y):
    return ((np.array(predict_result) - np.array(y))**2).mean()

# 예측값 리스트
predict_result = []

# 모든 x값을 한 번씩 대입하여 예측값 리스트 완성
for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("공부시간: %.f, 실제점수: %.f, 예측점수: %.f" % (x[i], y[i], predict(x[i])))

# a, b가 위와 같이 주어졌을 때의 MSE
print("MSE 최종값: ", mse_val(predict_result,y))
```
